```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('greyson-joralemon-9IBqihqhuHc-unsplash.jpg')
img.shape
```

```python
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(img[25, 25])
```

```python
df = pd.read_csv('wikipedia_color_names.csv')
df.head(10)
```

```python
def R_G_B(event, x, y, flags, params):
    minimum = 765
    for i in range(865):
        r = img[y, x][2] - df['Red (8 bit)'][i]
        g = img[y, x][1] - df['Green (8 bit)'][i]
        b = img[y, x][0] - df['Blue (8 bit)'][i]
        total = abs(r) + abs(g) + abs(b) 
        if total>=0:
            if total <= minimum:
                minimum = total
                index = df['Hex (24 bit)'][i]
                index2 = df['Name'][i]
                rgb = df[['Red (8 bit)', 'Green (8 bit)', 'Blue (8 bit)']].values[i]

    img_with_text = img.copy()
    text = f'code: {index}'
    cv2.putText(img_with_text, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    cv2.putText(img_with_text, f'{index2}', (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    cv2.imshow("image", img_with_text)

cv2.imshow('image', img)
cv2.setMouseCallback('image', R_G_B)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
